```python
# Writing a function for the relationship between x and y
# the goal is to use machine learning to figure out this relationship later

def solve_xy(x):
  y = (2 * x) - 1
  return y
```

```python
solve_xy(10)
```

```python
# Importing the required librariries
import tensorflow as tf
import numpy as np
from tensorflow import keras

print(tf.__version__)
```

```python
# Build a simple sequential model

model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
# Compile the model
model.compile(optimizer="sgd", loss="mean_squared_error")
```

```python
# Declare model inputs and outputs for the training
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
# Train the model
model.fit(xs, ys, epochs=500)
```

```python
# Make a prediction
print(model.predict([10]))
```
